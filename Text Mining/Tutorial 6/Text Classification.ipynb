```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Example sentence
sentence = "I don't like this movie"

# Preprocessing
tokens = nltk.word_tokenize(sentence.lower())

# Negation handling
negation_words = set(['no', 'not', "n't", 'never'])  # Add more negation words as needed
negated = False
for i, word in enumerate(tokens):
    if word in negation_words:
        negated = True
    elif negated and any(char in '.,;:?!' for char in word):
        negated = False
    elif negated:
        tokens[i] = 'NOT_' + word

# Sentiment Analysis
sid = SentimentIntensityAnalyzer()
sentiment_score = sid.polarity_scores(' '.join(tokens))

print("Sentiment Score:", sentiment_score)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline

# Load the DataFrame
df = pd.read_csv('train.csv')
# Preprocess the text data
vectorizer = TfidfVectorizer(stop_words="english")
X = vectorizer.fit_transform(df['reviews_content'])

# Load a pre-trained sentiment analysis model
model = make_pipeline(TfidfVectorizer(stop_words="english"), SVC(kernel='linear', probability=True))

# Replace 'positive' with 1 and 'negative' with 0 in the 'category' column
df['category'] = df['category'].map({'positive': 1, 'negative': 0})

Y = df['category']
# Fit the model
model.fit(X, Y)

# Perform prediction
prediction = model.predict(vectorizer.transform(df['reviews_content']))
print("Predicted category:", prediction)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline

df_train = pd.read_csv('train.csv')

# Convert category column to numeric
# df_train['category'] = df_train['category'].map({'positive': 1, 'negative': 0})

# Preprocess the text data
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df_train['reviews_content'])

print(X.shape)
# Initialize Logistic Regression model
logreg_model = LogisticRegression()

# Train the model
logreg_model.fit(X, df_train['category'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, df_train['category'], test_size=0.2)

# Initialize Logistic Regression model
logreg_model = LogisticRegression()

# Train the model
logreg_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = logreg_model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
